## time series for one group

look at all the rsvps. this way we could analyze
group memebership an cohorts

```python
id_group = "5027412"

sql = db_cursor.mogrify("""
            select events.id_event, events.time, 
            events.yes_rsvp_count, id_member, response from  rsvps
            left join events  using(id_event) 
            where events.id_group = %(id_group)s
            order by events.time
            """, {"id_group": id_group})
print sql
df_rsvps = pd.read_sql(sql, db_conn)
df_rsvps
```
