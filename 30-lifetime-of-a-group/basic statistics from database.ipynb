## How many events do we have data on?

```python
db_cursor.execute("select count(*) as count from events")
no_events = db_cursor.fetchone()['count']

db_cursor.execute("select count(distinct id_group) as count from events")
no_groups  = db_cursor.fetchone()['count']

db_cursor.execute("select count(*) as count from events left join rsvps using (id_event)")
no_rsvps  = db_cursor.fetchone()['count']



print "We have information on %d events from %d different groups in this database, with %d rsvps." % (no_events, no_groups, no_rsvps)
```

## How many groups do we have data on?

```python
df_group_stats = pd.read_sql("""
   select 
      events.id_group,    
      max(groups.name) as name_group,
      max(groups.join_mode) as join_mode,
      max(groups.no_members) as no_members, 
      min(groups.created) as created,
      max(groups.visibility) as visibility,
      min(events.time) as first_event,
      max(events.time) as last_event,      
      count(yes_rsvp_count) as no_events, 
      max(yes_rsvp_count) as max_attendees
   from events 
   left join groups using(id_group)
   group by id_group
   """, db_conn)
df_group_stats
```

## How many rsvps do we have data on?

```python
db_cursor.execute("select count(*) as count from rsvps")
no_rsvps = db_cursor.fetchone()['count']

db_cursor.execute("select count(distinct id_event) as count from rsvps")
no_events_with_rsvps = db_cursor.fetchone()['count']

print "We have %d rsvps for %d distinct events in this database" % (no_rsvps, no_events_with_rsvps)
```
