## time series for one group

look at all the rsvps. this way we could analyze
group memebership an cohorts

```python
id_group = "1338658"

sql = db_cursor.mogrify("""
            select events.id_event, events.time, 
            events.yes_rsvp_count, id_member, response from  rsvps
            left join events  using(id_event) 
            where events.id_group = %(id_group)s
            order by events.time
            """, {"id_group": id_group})
print sql
df_rsvps = pd.read_sql(sql, db_conn)
df_rsvps
```

Just a check: do the rsvps we have add up to the same number given on the event?

```python
sql = db_cursor.mogrify("""
            select time, 
            yes_rsvp_count_from_rsvps,
            1 as number_of_events
            from events
            where events.id_group = %(id_group)s
            order by time
            """, {"id_group": id_group})
print sql
df_rsvps = pd.read_sql(sql, db_conn)
df_rsvps
```

```python
ts = df_rsvps.set_index('time').resample('m').fillna(0)
ts.plot(kind='barh')
```

```python
sql = db_cursor.mogrify("""
            select events.id_group,  time, 
            yes_rsvp_count_from_rsvps
            from events left join groups using(id_group)
            where number_of_events > 3
            order by id_group, time
            limit 1000
            """)
print sql
df_all_rsvps = pd.read_sql(sql, db_conn)
#df_all_rsvps.set_index("time", inplace=True)
df_all_rsvps.head()
```
