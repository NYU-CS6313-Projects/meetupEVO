```python
%matplotlib inline
from glob import glob
import os
import random
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
def flatten_col(df, col):
    #print "flattening %s into df" % col
    try:
        sub_df = df[col].apply(pd.Series)
        sub_cols = list(sub_df.columns.values)
        #print sub_cols
        new_df = pd.merge(df, sub_df,  left_index=True, right_index=True, suffixes = ('_main','_%s' % col))
        rename_columns = {}
        for c in sub_cols:
            if c.startswith('%s_' % col):
                c_new =  c.replace('%s_' % col, '') + "_" + col
                rename_columns[c] = c_new
                # print "renaming %s to %s" % (c, c_new)
            elif c.endswith('_%s' % col):
                pass
            else:
                c_new = c + "_" + col
                rename_columns[c] = c_new
                #print "renaming %s to %s" % (c, c_new)

        new_df.rename(columns=rename_columns, inplace=True)
        del(new_df[col])
        return new_df
    except Exception as e:
        print "could not convert! %s" % e
        return df
```

```python
data_dir = "../10-data"
```

```python
random_file_sample = random.sample( glob("%s/events_updated/*.json" % data_dir), 5 )
print random_file_sample
```

```python
plt.title("Group Evolution by Number of Event")
for filename in random_file_sample:
    if os.path.getsize(filename)==0:
        continue
    df = pd.read_json(filename)
    if len(df) == 0:
        continue
        # ignore empty json
    df[u'filename'] = filename.replace(".json","").replace(data_dir+"/events_updated/","")
    df = flatten_col(df, "group")
    df = flatten_col(df, "rating") 
    plt.plot(df["yes_rsvp_count"],  label=df["name_group"][0])
    
plt.legend(loc=2)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10,6)

plt.show()
```
